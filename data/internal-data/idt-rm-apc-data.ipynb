```python
# This notebooks get two dataframes from the mobile_loan_applications collection.
# A dataframe have the details apc of the customers. *apc_detail_df
# The other dataframe have the score and a analysis of the user's acp. *apc_analyzed_score

import pandas as pd
from data import MOBILE_LOAN_DF, export_to_csv
```

```python
# Query to collections
df_mobile_loan = MOBILE_LOAN_DF(['_id', 'apc','approved'])
print('Ready')
```

```python
# Stores the loan_mobile IDs
id_lma_list, score_id_list = [], []

# List of apc score dataframe
score_customer_name_list, score_customer_last_name_list, score_personal_id_list = [], [], []
score_list, pi_list, exclusion_list = [], [], []
good_history_list, good_score_list, is_defaulter_apc, is_defaulter_60, competition_loan_list = [], [], [], [], []

# List of apc details dataframe
customer_name_list, customer_last_name_list,personal_id_list = [], [], []
history_id_list, institution_list, is_defaulter, type_list, balance_list, day_behind_list,\
approved_list = [], [], [], [], [], [], []



for ind in df_mobile_loan.index:
    id_lma = df_mobile_loan['_id'][ind]

    # Validates if in 'apc' key isn't empty.
    if not pd.isna(df_mobile_loan['apc'][ind]):
        # Try to get the 'Generales' field. Sometimes, it doesn't exist.
        try:
            customer_name = df_mobile_loan['apc'][ind]['apc']['Generales'][0]['NOMBRE'].title()
            customer_last_name = df_mobile_loan['apc'][ind]['apc']['Generales'][0]['APELLIDO'].title()
            personal_id = df_mobile_loan['apc'][ind]['apc']['Generales'][0]['IDENT_CLIE']
            approved = df_mobile_loan['approved'][ind]
        except:
            # When i want to get 'Generales' value, it is None or it's a empty list.
            customer_name =  ''
            customer_last_name = ''
            personal_id = ''
        finally:
            score_customer_name_list.append(customer_name)
            score_customer_last_name_list.append(customer_last_name)
            score_personal_id_list.append(personal_id)
            score_id_list.append(id_lma)
            # Gets score data
            try:
                score_list.append(df_mobile_loan['apc'][ind]['apc']['Score'][0]['SCORE'])
                pi_list.append(df_mobile_loan['apc'][ind]['apc']['Score'][0]['PI'])
                exclusion_list.append(df_mobile_loan['apc'][ind]['apc']['Score'][0]['EXCLUSION'])
            except:
                # When i want to get 'Score' value, it is None or it's a empty list.
                score_list.append('')
                pi_list.append('')
                exclusion_list.append('')

            # Gets Analyze apc data
            try:
                good_history_list.append(df_mobile_loan['apc'][ind]['apc']['AnalyzedApc']['buenHistorial'])
                good_score_list.append(df_mobile_loan['apc'][ind]['apc']['AnalyzedApc']['buenScore'])
                is_defaulter_apc.append(df_mobile_loan['apc'][ind]['apc']['AnalyzedApc']['isMoroso'])
                is_defaulter_60.append(df_mobile_loan['apc'][ind]['apc']['AnalyzedApc']['isMorosoEnUltimos60Dias'])
                competition_loan_list.append(df_mobile_loan['apc'][ind]['apc']['AnalyzedApc']['prestamoDeCompetencia'])
            except:
                # When i want to get 'AnalyzedApc' value, it is None or it's a empty list.
                good_history_list.append('')
                good_score_list.append('')
                is_defaulter_apc.append('')
                is_defaulter_60.append('')
                competition_loan_list.append('')

            try:
                # Try to get the different institutions of that customer. Again, sometimes it isn't with the regular format.
                for item in df_mobile_loan['apc'][ind]['apc']['Detalle']:

                    institution_list.append(item['NOM_ASOC'])
                    type_list.append(item['DESCR_CORTA_RELA'])
                    balance_list.append(item['SALDO_ACTUAL'])
                    day_behind_list.append(item['NUM_DIAS_ATRASO'])
                    history_id_list.append(item['HISTORIA'])

                    customer_name_list.append(customer_name)
                    customer_last_name_list.append(customer_last_name)
                    personal_id_list.append(personal_id)
                    approved_list.append(approved)
                    id_lma_list.append(id_lma)
            except:
                # When i want to get 'Detalle' value, it is None or it's a empty list.
                print(df_mobile_loan['apc'][ind])

# Create a new dataframes in this case.
apc_detail_df = pd.DataFrame(data={
    'id': id_lma_list,
    'personal_id':personal_id_list,
    'institution':institution_list,
    'history_id':history_id_list,
    'type':type_list,
    'firstname': customer_name_list,
    'lastname':customer_last_name_list,
    'balance':balance_list,
    # 'is_defaulter':is_defaulter,
    'day_behind':day_behind_list,
    'approved': approved_list
})
apc_analyzed_score =  pd.DataFrame(data={
    'id': score_id_list,
    'personal_id':score_personal_id_list,
    'firstname': score_customer_name_list,
    'lastname':score_customer_last_name_list,
    # These are Analyzed APC values
    'good_history':good_history_list,
    'good_score':good_score_list,
    'is_defaulter':is_defaulter_apc,
    'is_defaulter_60':is_defaulter_60,
    'competition_loan':competition_loan_list,
    # These are Score APC values
    'score': score_list,
    'pi': pi_list,
    'exclusion': exclusion_list
})
```

```python
# Export all dataframe to csv
export_to_csv(apc_detail_df, 'idt-apc-detail')
export_to_csv(apc_analyzed_score, 'idt-apc-analyze')
```
